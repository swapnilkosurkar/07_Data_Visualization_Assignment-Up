#Question 5. Show a pie chart to represent the percentage distribution of different sections in the array `sections`.

sections = ['Section A', 'Section B', 'Section C', 'Section D']

sizes = [25, 30, 15, 30]

#Answer:
Below is the code:

```python
import matplotlib.pyplot as plt

# Define the sections and their corresponding sizes
sections = ['Section A', 'Section B', 'Section C', 'Section D']
sizes = [25, 30, 15, 30]

# Define colors for each section
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']

# Highlight the first section
explode = (0.1, 0, 0, 0)

# Create the pie chart
plt.pie(sizes, explode=explode, labels=sections, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)

# Ensure the pie chart is a circle
plt.axis('equal')

# Add a title
plt.title('Percentage Distribution of Different Sections')

# Display the chart
plt.show()
```

This code will generate a pie chart with the specified sections and sizes. The explode parameter is used to highlight Section A by slightly pulling it out from the pie. The autopct parameter displays the percentage value on each slice.

#Question 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

#Answer:
Below is the code:

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set the random seed for reproducibility
np.random.seed(42)

# Generate random data
data_size = 100
X1 = np.random.normal(0, 1, data_size)
X2 = 0.5 * X1 + np.random.normal(0, 1, data_size)  # Correlated with X1
X3 = -0.2 * X1 + np.random.normal(0, 1, data_size)  # Negatively correlated with X1
X4 = np.random.normal(0, 1, data_size)  # Not correlated with X1

# Create a DataFrame
df = pd.DataFrame({'Feature1': X1, 'Feature2': X2, 'Feature3': X3, 'Feature4': X4})

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Set the style of the visualization
sns.set(style='white')

# Create a heatmap
plt.figure(figsize=(8, 6))
heatmap = sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, square=True, linewidths=.5, cbar_kws={"shrink": .5})

# Display the heatmap
plt.title('Correlation Matrix Heatmap')
plt.show()
```

This code will generate a synthetic dataset with correlated features and visualize the correlation matrix using a heatmap.

#Answer:

```python
import numpy as np
import pandas as pd
import plotly.express as px

# Set seed for reproducibility
np.random.seed(20)

# Create the dataset
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}

df = pd.DataFrame(data)

# Reshape the data for the heatmap
heatmap_data = df.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='mean')

# Generate the heatmap
fig = px.imshow(heatmap_data,
                labels={'x': "Month", 'y': "Day", 'color': "Sales"},
                x=['Jan', 'Feb', 'Mar', 'Apr', 'May'],
                y=list(range(1, 31)),
                color_continuous_scale='Viridis')

fig.update_layout(title='Monthly Sales Heatmap', xaxis_nticks=36)
fig.show()
```

```python
import numpy as np
import plotly.graph_objects as go

# Define the range for x and y
x_values = np.linspace(-5, 5, 100)
y_values = np.linspace(-5, 5, 100)
x_grid, y_grid = np.meshgrid(x_values, y_values)

# Compute the z values based on the function
z_values = np.sin(np.sqrt(x_grid**2 + y_grid**2))

# Create the surface plot
surface_plot = go.Figure(data=[go.Surface(z=z_values, x=x_grid, y=y_grid)])

# Customize the layout for better presentation
surface_plot.update_layout(
    title='3D Surface Plot of Z = sin(sqrt(x^2 + y^2))',
    autosize=False,
    width=700, height=700,
    margin=dict(l=65, r=50, b=65, t=90)
)

# Display the plot
surface_plot.show()
```

```python
from bokeh.plotting import figure, show
import numpy as np

# Generate an array of x values ranging from 0 to 10
x = np.linspace(0, 10, 100)
# Calculate the corresponding y values using the sine function
y = np.sin(x)

# Create a new Bokeh figure with a title and labels for the axes
sine_plot = figure(title="Sine Wave", x_axis_label='X-axis', y_axis_label='Y-axis')

# Add a line to the plot with a specified thickness and a legend entry
sine_plot.line(x, y, legend_label="sin(x)", line_width=2)

# Render the plot in the browser
show(sine_plot)
```

#Question 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

#Answer:

Below is the code:

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Enable inline plotting in Jupyter Notebook
output_notebook()

# Create random dataset
np.random.seed(42)
x_values = np.random.rand(100)
y_values = np.random.rand(100)
point_sizes = np.random.randint(10, 50, size=100)
point_colors = np.random.choice(['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd'], size=100)

# Prepare data for plotting
data_source = ColumnDataSource(data=dict(x=x_values, y=y_values, sizes=point_sizes, colors=point_colors))

# Set up the scatter plot
plot = figure(title="Scatter Plot with Random Values", x_axis_label='X Axis', y_axis_label='Y Axis')
plot.scatter('x', 'y', size='sizes', color='colors', source=data_source, alpha=0.6)

# Render the plot
show(plot)
```

```python
from bokeh.plotting import figure, output_file, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Prepare the data
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Create the figure
output_notebook()
p = figure(x_range=fruits, height=350, title="Fruit Counts",
           toolbar_location=None, tools="")

# Add bars
p.vbar(x='fruits', top='counts', width=0.9, source=source)

# Style the plot
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.yaxis.axis_label = "Count"

# Display the plot
show(p)
```

#Answer:

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Generate random data
data_hist = np.random.randn(1000)

# Calculate histogram
hist, edges = np.histogram(data_hist, bins=30)

# Enable output in notebook
output_notebook()

# Create a figure for the plot
p = figure(title="Data Distribution Histogram", background_fill_color="#fafafa")

# Add bars to the plot
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
       fill_color="blue", line_color="white", alpha=0.7)

# Customize the plot
p.y_range.start = 0
p.xaxis.axis_label = 'Data Values'
p.yaxis.axis_label = 'Count'
p.grid.grid_line_color="white"

# Display the plot
show(p)
```

#Question 5.Create a Bokeh heatmap using the provided dataset.


data_heatmap = np.random.rand(10, 10)

x = np.linspace(0, 1, 10)

y = np.linspace(0, 1, 10)

xx, yy = np.meshgrid(x, y)

#Answer:

Below is the code:


```python
import numpy as np
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColorBar, LinearColorMapper
from bokeh.palettes import Viridis256

# Provided dataset
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Flatten the data for plotting
x_flat = xx.flatten()
y_flat = yy.flatten()
data_flat = data_heatmap.flatten()

# Create a color mapper
color_mapper = LinearColorMapper(palette=Viridis256, low=data_flat.min(), high=data_flat.max())

# Create the figure
p = figure(title="Heatmap Example", x_axis_label='X', y_axis_label='Y',
           x_range=(0, 1), y_range=(0, 1), tools="hover", tooltips=[("value", "@image")])

# Add the heatmap
p.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1, color_mapper=color_mapper)

# Add color bar
color_bar = ColorBar(color_mapper=color_mapper, location=(0, 0))
p.add_layout(color_bar, 'right')

# Output to notebook
output_notebook()
show(p)
```

This code will create a heatmap using the data_heatmap array and display it using Bokeh. The ColorBar will provide a reference for the color mapping.
